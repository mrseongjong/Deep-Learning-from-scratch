```python
# 텍스트 파일 경로
text_file_path = 'test.txt'

try:
    # 파일을 열고 읽기 모드로 읽어옵니다.
    with open(text_file_path, 'r') as text_file:
        # 파일 내용을 읽어옵니다.
        file_content = text_file.read()
        
        # 읽어온 파일 내용을 출력합니다.
        print(file_content)

except FileNotFoundError:
    print(f"파일을 찾을 수 없습니다: {text_file_path}")
except Exception as e:
    print(f"파일을 읽는 도중 오류가 발생했습니다: {str(e)}")

```

#### Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

# data
x = np.arange(0, 22, 0.1)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
# data
x = np.arange(0, 22, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1)
plt.plot(x,y2)

# plotting a graph
plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle='--', label='cos', color='black') # dashed line
plt.xlabel('x') # x-axis
plt.ylabel('y') # y-axis
plt.title('sin & cos') # title
plt.legend() # label box
plt.show()
```
